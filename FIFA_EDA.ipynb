# 4.Who is the tallest player in FIFA?

```python
# Define a function to convert height from string format (feet'inches) to inches
def height_to_inches(height):
    try:
        feet, inches = map(int, height.split("'"))
        return feet * 12 + inches  # Convert feet to inches and add inches
    except (AttributeError, ValueError):
        return float('nan')  # Handle invalid or non-string data

# Convert height to inches
data['Height_inches'] = data['Height'].apply(height_to_inches)

# Find the tallest player
tallest_height = data['Height_inches'].max()
tallest_players = data[data['Height_inches'] == tallest_height]
tallest_player_name = tallest_players.iloc[0]['Name']
tallest_player_height = tallest_players.iloc[0]['Height']

print("Tallest player in FIFA:", tallest_player_name)
print("Height:", tallest_player_height)
```

# 5.Plot a histogram to get the salary range of the players.

```python
import matplotlib.pyplot as plt

# Plot a histogram of the salary range
plt.hist(data['Wage'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Wage')
plt.ylabel('Value')
plt.title('Salary Range of Players')
plt.grid(True)
plt.show()
```

# 6. Which club has the most number of players?

```python
most_players_club = data['Club'].value_counts().sort_values(ascending=False).index[0]
print("Club with the most number of players:", most_players_club)
```
